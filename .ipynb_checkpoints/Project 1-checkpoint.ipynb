```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as sts
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

```

```python
# Read Files 
Dataset=pd.read_csv("DATA/VSRR_-_State_and_National_Provisional_Counts_for_Live_Births__Deaths__and_Infant_Deaths.csv")
PopData=pd.read_csv("DATA/nst-est2019-01.csv", header=3)
Dataset.head()
```

```python
#formatting the data to start as certain row and rename location header
PopData.drop(PopData.index[[0,1]])

PopData = PopData.rename(columns={"Unnamed: 0":"State"})

PopData.head()
```

```python
#Droping Indicator not equal to "Number of Live Births"

Data_filter = Dataset[Dataset['Indicator'] == "Number of Live Births"]
Data_filter2 = Data_filter[Data_filter['Year'] == 2018]
Data_filter3 = Data_filter2[Data_filter2['Period'] == 'Monthly']

#Validating
#Data_filter3["Year"].unique()
#Data_filter3["Indicator"].unique()
#Data_filter3["Period"].unique()

Data_filter3
```

```python
#Rename the Data Value column to a more relevant name of Live Births

Data_filter3 = Data_filter3.rename(columns={"Data Value":"Live Births"})
Data_filter3
```

```python
#Groupby state and sum values to get a full year's live births

Grpby_State= Data_filter3.groupby(["State"])
#Grpby_State.get_group('ALABAMA')
Grpby_State.head()
```

```python
#Sum the Live births by state for full year 2018 

FY_Births_State= Grpby_State["Live Births"].sum()
FY_Births_State
```

```python
#bring in second data file and clean it up to only show State and 2017 population data. drop all other columns
#Convert state names to upper case to be able to merge the 2 datasets by state. 

NewPop2017 = PopData.loc[:,['State', '2017']]
NewPop2017_df = pd.DataFrame(NewPop2017)
Pop2017=NewPop2017_df.dropna(how="any")

Pop2017["State"]=Pop2017["State"].str.upper()
Pop2017.head()
```

```python
#Merge the two datasets on State. 

merge_data = pd.merge(FY_Births_State, Pop2017, on="State")
merge_data
```

```python
#Replace comma in 2017 data
merge_data["2017"]=merge_data["2017"].str.replace(",","")
merge_data.head()
```

```python
#extract digits from string
merge_data["2017"]=merge_data["2017"].str.extract('(^\d*)')
merge_data["2017"]=merge_data["2017"].astype(int)
```

```python
#The formula for crude birth rate is: CBR = (b ÷ p) X 1,000.
#crude birth rate (CBR),  number of live births (b), total population (p)
birth=merge_data["Live Births"]
population=merge_data["2017"]
birthrate=(birth/population)*1000
birthrate.map("{:.2f}".format)
```

```python
#Print the merge data
merge_data["Birthrate/1000"]=birthrate
merge_data
```

```python
#Plotting the birthrate by state for us 2018 

x_axis = merge_data['State']
y_axis = merge_data['Birthrate/1000']
plt.figure(figsize=(20,10))
plt.bar(x_axis, y_axis, color='r', alpha=0.5, align="center")
plt.ylim(0, max(y_axis)+1)
plt.xticks(merge_data["State"], rotation="vertical")
plt.title("Birthrate by State in 2018")
plt.xlabel("State")
plt.ylabel("Birthrate per 1000 people")
plt.savefig("Birthrate Values by State 2018.png")
plt.show()
```

```python
# Dual Axis Graph to compare values and rate of live birth by state

Data_minus_US = merge_data[merge_data['State'] != 'UNITED STATES']
fig,ax = plt.subplots(figsize=(15,10))
x_axis = Data_minus_US['State']
y_axis_2 = Data_minus_US['Birthrate/1000']
y_axis_1 = Data_minus_US['Live Births']
ax2 = ax.twinx()  # instantiate a second axes that shares the same y-axis
ax.bar(x_axis, y_axis_1, color='r', alpha=0.5, align="center")
ax2.plot(x_axis, y_axis_2, marker="o", color='b', alpha=0.5)
ax2.set_ylim(0, max(y_axis_2)+ 1)
ax.set_xticks(Data_minus_US["State"])
ax.tick_params('x', labelrotation=90)
ax.set_xlabel('States')
ax.set_ylabel('# of live births')
ax2.set_ylabel('Birthrate per 1000 people')
ax2.set_title('Number of Live Births vs. Birthrate by State 2018')
plt.savefig("Number of Live Births vs. Birthrate by State 2018.png")
plt.show()
```

```python
#Plot a Histogram to determine the average and the outliers
plt.hist(Data_minus_US['Birthrate/1000'])
plt.xlabel('# of States')
plt.ylabel('Birth rate per 1000')
plt.savefig("histogram.png")
plt.show()
print(sts.normaltest(Data_minus_US['Birthrate/1000'].sample(50)))
#ploting
quartiles = Data_minus_US['Birthrate/1000'].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
print(f"The lower quartile of birthrate is: {lowerq}")
print(f"The upper quartile of birthrate is: {upperq}")
print(f"The interquartile range of birthrate is: {iqr}")
print(f"The the median of birthrate is: {quartiles[0.5]} ")
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
state_birth_rate = Data_minus_US.loc[(Data_minus_US['Birthrate/1000'] < lower_bound) | (Data_minus_US['Birthrate/1000'] > upper_bound)]

state_birth_rate
```

```python
#Create a DF for only the United States by Month for live births

Data_US_by_Mo = Data_filter3[Data_filter3['State'] == 'UNITED STATES']
Data_US_by_Mo
```

```python
#plot US by month 

x_axis = Data_US_by_Mo['Month']
y_axis = Data_US_by_Mo['Live Births']
plt.figure(figsize=(20,10))
plt.bar(x_axis, y_axis, color='r', alpha=0.5, align="center")
plt.ylim(0, max(y_axis)+100000)
plt.xticks(Data_US_by_Mo["Month"], rotation="vertical")
plt.title("No. of live births in the US by month in 2018")
plt.xlabel("Month")
plt.ylabel("Number. of Live Births in US by Month 2018")
plt.savefig("Number. of Live Births in US by Month 2018.png")
plt.show()
```

```python
#Merge Feb and August (Highest and Lowest birthmonths) into one DF 

DataFilter_minus_US= Data_filter3[Data_filter3['State'] != 'UNITED STATES']

Low_month= DataFilter_minus_US[DataFilter_minus_US['Month']=="February"]
#Btm_month

High_month= DataFilter_minus_US[DataFilter_minus_US['Month']=="August"]
High_month

FebAug_df = pd.merge(Low_month, High_month, on="State", how="left")
FebAug_df
```

```python
#plot an overlay graph of the highest and lowest birth month ( Feb/Aug) 
x_axis = Low_month['State']
y_axis_Feb = Low_month['Live Births']
y_axis_Aug = High_month['Live Births']
plt.figure(figsize=(20,10))
plt.bar(x_axis, y_axis_Feb, color='green', alpha=0.5, align="center", label='February')
plt.bar(x_axis, y_axis_Aug, color='black', alpha=0.5, align="center", label= 'August')
plt.ylim(0, max(y_axis_Aug)+10000)
plt.xticks(DataFilter_minus_US["State"], rotation="vertical")
plt.title("Top Month vs Lowest Month by State Comparison in 2018")
plt.xlabel("State")
plt.ylabel("Live Births Feb & Aug")
plt.legend(loc="best", prop={'size':15})
plt.savefig("Top Month vs Lowest Month by State Comparison in 2018.png")
plt.show()
```

```python
#Cal the discrepancy between the highest and lowest month for all states. 

Pct_Diff= (FebAug_df["Live Births_y"]-FebAug_df["Live Births_x"])/FebAug_df["Live Births_x"] *100

FebAug_df["Percent Change"]=Pct_Diff
FebAug_df
```

```python
#Plot the percent change by state for highest/lowest month

x_axis = FebAug_df['State']
y_axis = FebAug_df['Percent Change']

plt.figure(figsize=(20,10))
plt.plot(x_axis, y_axis, color='red', marker="o", alpha=0.5)
#plt.ylim(0, max(y_axis)+5)
plt.xticks(FebAug_df["State"], rotation="vertical")
plt.title("Top Month vs Lowest Month Percent Chg by State (2018)")
plt.xlabel("State")
plt.ylabel("Percent Chnage")
plt.savefig("Top Month vs Lowest Month Percent Chg by State 2018.png")
plt.show()
```
